```python
# Calculator in which 4 arithmatic operations are possible with already fixed numbers.
#And operation stop after one calculation.

first_num= 15
second_num=28

operation= input("Use *, +, /, - for arithmetic")

if operation == "+":
  print(first_num + second_num)

elif operation == "-":
  print(first_num - second_num)

elif operation == "*":
  print(first_num * second_num)

elif operation == "/":
  print(first_num / second_num)

else:
  print("Invalid operation")
```

```python
# Calculator in which 4 arithmatic operations are possible with already fixed numbers.
#And operation remain continued.

while True:
  first_num= 15
  second_num=28

  operation= input("Use *, +, /, - for arithmetic")

  if operation == "+":
    print(first_num + second_num)

  elif operation == "-":
    print(first_num - second_num)

  elif operation == "*":
    print(first_num * second_num)

  elif operation == "/":
    print(first_num / second_num)

  else:
    print("Invalid operation")
    break
```
